Crie uma classe chamada Funcionario que tenha os atributos privados __nome e __salario_base. Crie métodos getter para eles. Crie um método chamado calcular_pagamento() que simplesmente retorne o salário base. Crie uma classe filha chamada Professor que herde de Funcionario . O Professor possui um atributo adicional chamado horas_extras_orientacao. Aplique o Polimorfismo: sobrescreva o método calcular_pagamento() na classe Professor para que ele retorne o salário base somado a (horas_extras * 50 reais).

```python
class Funcionario:
    def __init__(self, nome, salario_base):
        self.__nome = nome
        self.__salario_base = salario_base

    def get_nome(self):
        return self.__nome

    def get_salario_base(self):
        return self.__salario_base

    def calcular_pagamento(self):
        return self.__salario_base
```

```python
class Professor(Funcionario):
    def __init__(self, nome, salario_base, horas_extras_orientacao):
        super().__init__(nome, salario_base)
        self.horas_extras_orientacao = horas_extras_orientacao

    def calcular_pagamento(self):
        return self.get_salario_base() + (self.horas_extras_orientacao * 50)
```

Testes

```python
funcionario = Funcionario("Zeca Pagodinho", 3000)
print(f"Funcionário: {funcionario.get_nome()}")
print(f"Salário base: R${funcionario.get_salario_base()}")
print(f"Pagamento: R${funcionario.calcular_pagamento()}")

print("\n" + "-"*30 + "\n")

professor = Professor("Jorge Aragão", 4000, 10)
print(f"Professor: {professor.get_nome()}")
print(f"Salário base: R${professor.get_salario_base()}")
print(f"Horas extras: {professor.horas_extras_orientacao}")
print(f"Pagamento: R${professor.calcular_pagamento()}")
```
