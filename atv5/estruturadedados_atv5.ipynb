Cada grupo irá desenvolver scripts para o processamento e salvamento uma matriz, depois carrega e manipula novamente os dados.

### 1a etapa: Criação da matriz
Gere uma matriz 5 x  5 de inteiros aleatórios no intervalo de 1 a 100 usando NumPy.

### 2a etapa: Salvamento em arquivo
Salve a matriz gerada em um arquivo de texto ou CSV

### 3a etapa: Leitura do arquivo
Carregue a matriz novamente do arquivo

### 4a etapa: Manipulação
Extraia a submatriz central 3 x 3, achate e ordene essa submatriz, e reinserir os valores ordenados no centro da matriz, mantendo as demais posições.

Calcule e salve em arquivo separado as estatísticas básicas: média, mediana, máximo e mínimo (tanto da matriz geral quanto da submatriz central).

Troque as linhas pares pelas ímpares e gere a transposta da matriz.

### 5a etapa: Salvamento final
Salve a matriz final (depois de todas as manipulações) em novo arquivo.

```python
import numpy as np
import csv

# Criação da matriz
matriz = np.random.randint(1, 101, size=(5, 5))
# print(matriz)

# Salvamento em arquivo
with open('matriz_original.csv', mode='w+', newline='') as f:
    escritor = csv.writer(f)
    escritor.writerows(matriz)

# Leitura do arquivo
matriz_carregada_lista = []
with open('matriz_original.csv', mode='r') as f:
    leitor = csv.reader(f)
    for linha in leitor:
        matriz_carregada_lista.append([int(item) for item in linha])

matriz_carregada = np.array(matriz_carregada_lista)

# Manipulação

# a)
submatriz = matriz_carregada[1:4, 1:4].copy()
submatriz_ordenada = np.sort(submatriz.flatten()).reshape(3, 3)

matriz_manipulada = matriz_carregada.copy()
matriz_manipulada[1:4, 1:4] = submatriz_ordenada

# b)
stats = {
    "geral": {
        "media": np.mean(matriz_manipulada),
        "mediana": np.median(matriz_manipulada),
        "max": np.max(matriz_manipulada),
        "min": np.min(matriz_manipulada)
    },
    "submatriz": {
        "media": np.mean(submatriz_ordenada),
        "mediana": np.median(submatriz_ordenada),
        "max": np.max(submatriz_ordenada),
        "min": np.min(submatriz_ordenada)
    }
}

with open('estatisticas.txt', 'w', encoding='utf-8') as f:
    for rotulo, valores in stats.items():
        f.write(f"{rotulo}: {valores}\n")

# c)
matriz_final = matriz_manipulada.copy()
matriz_final[[0, 1, 2, 3]] = matriz_final[[1, 0, 3, 2]]
matriz_transposta = matriz_final.T

# Salvamento final
with open('matriz_final.csv', mode='w', newline='') as f:
    escritor = csv.writer(f)
    escritor.writerows(matriz_transposta)
```
