Utilize o arquivo criado a partir do script a seguir:

```python
import csv

def gerar_dados():
    dados = [
        ['ID', 'Produto', 'Quantidade', 'Preço Unitário', 'Categoria'],
        [1, 'Teclado Mecânico', 15, 250.00, 'Periféricos'],
        [2, 'Monitor 24"', 5, 1200.00, 'Hardware'],
        [3, 'Mouse Gamer', 30, 180.00, 'Periféricos'],
        [4, 'Memória RAM 16GB', 12, 450.00, 'Hardware'],
        [5, 'Headset Wireless', 8, 350.00, 'Acessórios'],
        [6, 'Placa de Vídeo', 3, 3500.00, 'Hardware'],
        [7, 'Cabo HDMI', 50, 25.00, 'Acessórios'],
        [8, 'SSD 1TB', 10, 550.00, 'Hardware']
    ]

    with open('vendas.csv', 'w', newline='', encoding='utf-8') as arquivo:
        escritor = csv.writer(arquivo)
        escritor.writerows(dados)
    print("Arquivo 'vendas.csv' gerado com sucesso!")

gerar_dados()
```

A partir do arquivo criado desenvolver um script que processe dados de vendas e gere um relatório filtrado.
Instruções:

1. Leitura: Leia o arquivo vendas.csv.
2. Cálculos Estatísticos:
- Calcule a média dos preços unitários.
- Mostre a quantidade de produtos cadastrados.
- Identifique o produto com o maior valor de estoque total (Quantidade × Preço).
- Identifique o produto com o menor preço unitário.
3. Filtragem e Escrita: Crie um novo arquivo chamado produtos_acima.csv
- Este arquivo deve conter apenas os produtos da categoria "Hardware" com preço unitário acima de R$ 500,00.

```python
# from pprint import pprint

ID = 0
PRODUTO = 1
QUANTIDADE = 2
PRECO_UNITARIO = 3
CATEGORIA = 4

res = {
  'total_preco_unitario': 0,
  'quantidade_produtos': 0,
  'produto_maior_valor_estoque': None,
  'produto_menor_preco_unitario': None,
  'hardware_preco_maior_500': []
}

with open('vendas.csv', 'r', newline='', encoding='utf-8') as arquivo:
  tabela = list(csv.reader(arquivo))
  cabecalho = tabela.pop(0)

  for row in tabela:
    res['total_preco_unitario'] += float(row[PRECO_UNITARIO])
    res['quantidade_produtos'] += 1

    if not res['produto_maior_valor_estoque'] or \
      int(row[QUANTIDADE]) * float(row[PRECO_UNITARIO]) > \
      int(res['produto_maior_valor_estoque'][QUANTIDADE]) * \
      float(res['produto_maior_valor_estoque'][PRECO_UNITARIO]):
      res['produto_maior_valor_estoque'] = row

    if not res['produto_menor_preco_unitario'] or \
      float(row[PRECO_UNITARIO]) < \
      float(res['produto_menor_preco_unitario'][PRECO_UNITARIO]):
      res['produto_menor_preco_unitario'] = row

    if row[CATEGORIA] == 'Hardware' and float(row[PRECO_UNITARIO]) > 500.00:
        res['hardware_preco_maior_500'].append(row)

media_precos_unitarios = res['total_preco_unitario'] / res['quantidade_produtos'] if res['quantidade_produtos'] > 0 else 0
print("\n--- Relatório estatístico ---")
print(f"Média dos preços unitários: R$ {media_precos_unitarios:.2f}")
print(f"Quantidade de produtos cadastrados: {res['quantidade_produtos']}")
print(f"Produto com o maior valor de estoque total: {res['produto_maior_valor_estoque'][PRODUTO]} (R$ {float(res['produto_maior_valor_estoque'][PRECO_UNITARIO]):.2f})")
print(f"Produto com o menor preço unitário: {res['produto_menor_preco_unitario'][PRODUTO]} (R$ {float(res['produto_menor_preco_unitario'][PRECO_UNITARIO]):.2f})")

if res['hardware_preco_maior_500']:
  with open('produtos_acima.csv', 'w+', newline='', encoding='utf-8') as arquivo_saida:
    escritor_saida = csv.writer(arquivo_saida)
    escritor_saida.writerow(cabecalho)
    escritor_saida.writerows(res['hardware_preco_maior_500'])
  print("\nArquivo 'produtos_acima.csv' gerado com sucesso com os produtos filtrados!")
```
